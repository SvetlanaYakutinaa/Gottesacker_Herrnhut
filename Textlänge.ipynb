```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Verzeichnis, in dem sich die .txt-Dateien befinden
directory = 'Text/männlich'

# Variable, um den gesamten Inhalt der Dateien zu speichern
corpora_m = ''

# Durchlaufe alle Dateien im angegebenen Verzeichnis
for filename in os.listdir(directory):
    # Prüfe, ob die Datei eine .txt-Datei ist
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        # Öffne die Datei und lese ihren Inhalt
        with open(file_path, 'r', encoding='utf-8') as file:
            corpora_m += file.read() + '\n'
```

```python
# corpora_m
```

```python
#Average Textlength
corpora_m_split = corpora_m.split()
text_length_m = len(corpora_m_split)
average_text_length_m = text_length_m / 4
average_text_length_m
```

```python
#Average Sentence length
corpora_m_split_S = corpora_m.split(".")
sentence_length_m = len(corpora_m_split_S)
average_sentence_length_m = sentence_length_m / 4
average_sentence_length_m
```

```python
# Verzeichnis, in dem sich die .txt-Dateien befinden
directory = 'Text/weiblich'

# Variable, um den gesamten Inhalt der Dateien zu speichern
corpora_w = ''

# Durchlaufe alle Dateien im angegebenen Verzeichnis
for filename in os.listdir(directory):
    # Prüfe, ob die Datei eine .txt-Datei ist
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        # Öffne die Datei und lese ihren Inhalt
        with open(file_path, 'r', encoding='utf-8') as file:
            corpora_w += file.read() + '\n'
```

```python
#Average Textlength
corpora_w_split = corpora_w.split()
text_length_w = len(corpora_w_split)
average_text_length_w = text_length_w / 4
average_text_length_w
```

```python
#Average Sentence length
corpora_w_split_S = corpora_w.split(".")
sentence_length_w = len(corpora_w_split_S)
average_sentence_length_w = sentence_length_w / 4
average_sentence_length_w
```

```python
# Visualisation text length

# Position der Balken auf der x-Achse
labels = ['Männer', 'Frauen']
values = [average_text_length_m, average_text_length_w]

# Erstellen des Balkendiagramms
fig, ax = plt.subplots()
ax.bar(labels, values, color=['blue', 'purple'])

# Hinzufügen von Beschriftungen und Titel
ax.set_xlabel('Geschlecht')
ax.set_ylabel('Wörter')
ax.set_title('Durchschnittliche Textlänge')

# Anzeigen des Diagramms
plt.show()

# Gafik als png speichern
plt.savefig('Output/textlänge.png')
```

```python
# Visualisation sentence length

# Position der Balken auf der x-Achse
labels = ['Männer', 'Frauen']
values = [average_sentence_length_m, average_sentence_length_w]

# Erstellen des Balkendiagramms
fig, ax = plt.subplots()
ax.bar(labels, values, color=['blue', 'purple'])

# Hinzufügen von Beschriftungen und Titel
ax.set_xlabel('Geschlecht')
ax.set_ylabel('Sätze')
ax.set_title('Durchschnittliche Satzlänge')

# Anzeigen des Diagramms
plt.show()

# Gafik als png speichern
plt.savefig('Output/satzlänge.png')
```
